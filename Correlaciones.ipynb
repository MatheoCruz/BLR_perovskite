```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import seaborn as sns
```

```python
bd_grande_limitada = pd.read_csv("/content/drive/MyDrive/Deep Learning/TG/Preprocesamiento/BD_GRANDE_LIMITADA.csv")
df = pd.read_csv("/content/drive/MyDrive/Deep Learning/TG/Preprocesamiento/data_sin_nan.csv")
df.info()
```

```python

# Definir transformaciones
transformations = [None, np.log, np.exp, np.square, np.sqrt]

# Definir tipos de correlaci√≥n
correlation_types = ["pearson"]

# Calcular correlaciones
correlations = {}
for column_name in df.columns:
    column = df[column_name]
    column_correlations = {}
    for transform in transformations:
        column_transformed = column if transform is None else column.apply(transform)
        for correlation_type in correlation_types:
            df_transformed = pd.concat([df.drop(column_name, axis=1), column_transformed], axis=1)
            corr = df_transformed.corr(method=correlation_type)
            column_correlations[f"{correlation_type}_{str(transform)}"] = corr[column_name][:-1]
        correlations[column_name] = column_correlations

print(correlations)
```

```python
def Transformaciones(data):
  transformacion = {
      'A': np.exp,
      'B': np.exp,
      'X': np.exp,
      'Band_gap': np.exp,
      'DeltaH': np.log,
      'DeltaL': np.exp,
      'Grain_size': np.sqrt,
      'Thickness': np.log,
      'PCE': None,
      'Voc': np.exp,
      'Jsc': None,
      'FF': np.exp,
  }
  retorno = {}
  for column_name in data.columns:
    column = data[column_name]
    column_transformada = column if transformacion[column_name] is None else column.apply(transformacion[column_name])
    retorno[column_name] = column_transformada
  retorno = pd.DataFrame(retorno)
  return retorno
```
